```python
# import libraries
import numpy as np
import pickle

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# load model

# compaile= false  →  modeli derlemeden yükle
model=load_model("lstm_regression_yelp_model.h5",compile=False)
```

```python
# load tokenizer
with open("tokenizer.pkl","rb") as f:
    tokenizer=pickle.load(f)
```

```python
# örnek yorum

# 2 yıldız - 5 yıldız
text=[
    "After waiting for almost 30 minutes to trade in an old phone part of the buy back program, our customer service rep incorrectly processed the transaction. This led to us waiting another 30 minutes for him to correct it. Don't visit this store if you want pleasant or good service.",
    "Can't miss stop for the best Fish Sandwich in Pittsburgh.",
    "best experince ever",
    " I had a terrible experience at this restaurant. The service was slow, and the food was cold when it arrived. I will not be returning.",
    "Absolutely fantastic! The staff were friendly and attentive, and the food was delicious. Highly recommend to anyone looking for a great dining experience."
]
```

```python
# transform text to sequences
sequences=tokenizer.texts_to_sequences(text)

# pad sequences
padded=pad_sequences(sequences,maxlen=100,padding='post')
```

```python
# LSTM prediction
predictions=model.predict(padded)

# POST PROCESSING

# biz ilk veriyi 0-1 aralığına ölçeklendirmiştik, şimdi tekrar 1-5 aralığına getirelim
predictions_scaled=predictions*5 
```

```python
# write predictions
for i, review in enumerate(text):
    print(f"Review: {review}")
    print(f"Predicted Rating: {predictions_scaled[i][0]:.2f}")
    print()
```
